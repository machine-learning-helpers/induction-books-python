```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0.5
sigma = 0.5

# Logistic
distro_logistic = np.random.logistic (mu, sigma, 10000)
plt.hist (distro_logistic, 50, normed = True)

# Normal
distro_normal = np.random.normal (mu, sigma, 10000)
plt.hist (distro_normal, 50, normed = True)

# Plot
plt.figure()
plt.show()
```

```python
# CDF
plt.hist (distro_logistic, 50, normed = True, cumulative = True)
plt.figure()
```
